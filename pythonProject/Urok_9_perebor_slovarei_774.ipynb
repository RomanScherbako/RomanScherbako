## defaultdict

```python
counter = {}

text2 = text.replace(",", '').replace(".", '').replace('-', '').replace('!', '').replace(';', '').replace('"', '').lower()
spisok_slov = text2.split()

for word in spisok_slov:
    if word in counter:
        counter[word] += 1
    else:
        counter[word] = 1
```

```python
from collections import defaultdict

text2 = text.replace(",", '').replace(".", '').replace('-', '').replace('!', '').replace(';', '').replace('"', '').lower()
spisok_slov = text2.split()

counter_dd = defaultdict(int)

for word in spisok_slov:
    counter_dd[word] += 1
```
